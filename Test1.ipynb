```python
inp = input()
proverb = inp.lower()[:-1].split(' ')
last = inp.lower()[-1]
for ind, word in enumerate(proverb):
  if len(word) > 1:
    n = 0
    new_word = ''
    while n != min(len(word[::2]), len(word[1::2])):
      new_word += word[1::2][n] + word[::2][n]
      n += 1

    if len(word) % 2 != 0:
      new_word += word[::2][n]
    else:
      pass
  
    proverb[ind] = new_word

print(' '.join(proverb).capitalize()+last)
```

```python
import math
numb = input()
cmd = input()

F = True
f = 10 ** 11

def check(numb):
    try:
        float(numb)
        return True
    except ValueError:
        return False
        

if check(numb):
    numb = float(numb)
    if float(numb) % 1 == 0:
        numb = int(numb)
else:
    print('None')
    F = False

def square(a):
  return (math.trunc((a**2) * f) / f)

def cube(a):
  return (math.trunc((a**3) * f) / f)

def si(a):
  return (math.trunc(math.sin(math.radians(a)) * f) / f)

def co(a):
  return (math.trunc(math.cos(a) * f) / f)

def absolute(a):
  return (math.trunc(abs(a) * f) / f)

if F:
    if cmd == 'Квадрат числа' and check(square(numb)):
      print(square(numb))
    elif cmd == 'Куб числа' and check(cube(numb)):
      print(cube(numb))
    elif cmd == 'Синус числа' and check(si(numb)):
      print(si(numb))
    elif cmd == 'Косинус числа' and check(co(numb)):
      print(co(numb))
    elif cmd == 'Модуль числа' and check(absolute(numb)):
      print(absolute(numb))
    else:
      print('None')
```
