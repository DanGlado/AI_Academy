### Методы класса

```python
class Car():
  def __init__(self, model, color, size):
    self.model = model
    self.color = color
    self.size = size

  def stop(self):
    print("Stop!")
    self.speed = 0

  def drive(self):
    print('WRRRUM!')
    self.speed = 100
```

```python
volvo = Car('Volvo', 'green', 9)
volvo.drive()
print(f'Скорость {volvo.model} равна {volvo.speed} км/ч')
volvo.stop()
print(f'Скорость {volvo.model} равна {volvo.speed} км/ч')
```

```python
class Car():
  def __init__(self, model, color, size):
    self.model = model
    self.color = color
    self.size = size

  def stop(self):
    print("Stop!")
    self.speed = 0
    print(f'Скорость {self.model} равна {self.speed} км/ч')

  def drive(self, speed):
    print('WRRRUM!')
    self.speed = speed
    print(f'Скорость {self.model} равна {self.speed} км/ч')
```

### Инкапсуляция

```python
class Car():
  def __init__(self, model, color, size):
    self.model = model
    self.color = color
    self.size = size
    self.__num_wheels = 4

  def stop(self):
    print("Stop!")
    self.speed = 0
    print(f'Скорость {self.model} равна {self.speed} км/ч')

  def drive(self, speed):
    print('WRRRUM!')
    self.speed = speed
    print(f'Скорость {self.model} равна {self.speed} км/ч')
```

### Наследование

```python
class Transmission():
  def __init__(self):
    self.type= 'Manual'

  def first_speed(self):
    print('first speed')
```

```python
class Car( Transmission):
  def __init__(self, model, color, size):
    super().__init__()
    self.model = model
    self.color = color
    self.size = size
```

```python
kamaz = Car('Kamaz', 'orange', 15)
kamaz.first_speed()
kamaz.type
```

### Множественное наследование

```python
class Engine():
  def __init__(self):
    self.type = 'diesel'
```

```python
class Car(Engine, Transmission):
  def __init__(self, model, color, size):
    super().__init__()
    self.model = model
    self.color = color
    self.size = size
```

```python
kamaz = Car('Kamaz', 'orange', 15)
kamaz.first_speed()
kamaz.type
```

```python
class Car():
  def __init__(self, model, color, size):
    super().__init__()
    self.model = model
    self.color = color
    self.size = size
  
  def drive(self, speed):
    print('WRRRUM!')
    self.speed = speed
    print(f'Скорость {self.__class__} равна {self.speed} км/ч')
```

```python
class Boat():
  def __init__(self, model, color, size):
    super().__init__()
    self.model = model
    self.color = color
    self.size = size
  
  def drive(self, speed):
    print('Bulbul!')
    self.speed = speed
    print(f'Скорость {self.__class__} равна {self.speed} км/ч')
```

```python
audi = Car('Audi', 'black', 8)
yamaha = Boat('Yamaha', 'white', 10)

audi.drive(100)
yamaha.drive(50)
```
